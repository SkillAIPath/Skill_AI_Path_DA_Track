10. An analyst is exploring a dataset. They run `.info()` and see a 'revenue' column has a Dtype of 'object'. When they run `.describe()`, the 'revenue' column is missing from the output. What is the most likely conceptual issue with the 'revenue' column?
    - A. The numbers in the 'revenue' column are too large for `.describe()` to calculate statistics on.
    - B. The `pd.set_option()` command used for formatting floats has incorrectly hidden the column.
    - C. The column must be entirely empty or contain only null (NaN) values.
    - D. The column contains non-numeric values, such as currency symbols ('$') or commas, forcing pandas to treat it as text.

<details>
<summary>Click for answer</summary>
<pre>The answer is Option D.
<bold>Reason:</bold>
    This is the core concept. `.describe()` by default only provides statistics for numerical columns. If pandas sees even one non-numeric entry, it classifies the whole column as a generic 'object' (text), which is then excluded from the statistical summary.
</details>
